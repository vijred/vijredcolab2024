```python
import openai
from dotenv import load_dotenv
import os

load_dotenv()
os.environ["OPENAI_API_KEY"] = os.getenv("openai_api_key")

client = openai.OpenAI()
```

Define a function called **get_output_from_messages**.    
When called, this function will pass a list of messages to OpenAI, along wtih other parameters (many optional) that we set.  The function will return the completion generated by the call to the OpenAI LLM.

```python
def get_output_from_messages(messages_input, model_input = "gpt-3.5-turbo", temperature_input = 0):
    
    response = client.chat.completions.create(
        model = model_input,
        messages = messages_input,
        temperature = temperature_input,
    )

    response = response.choices[0].message.content
    
    return response
```

Create a messages object, using a list of Python disctionaries.
Call get_output_from_messages function, with messages and setting temperature equal to .5.

```python
messages =  [  
{'role':'system', 'content':'You are friendly customer service telephone operator.'},    
{'role':'user', 'content':'Hi, my name is Liza'}  ]
response = get_output_from_messages(messages, temperature_input=.5)
print(response)
```

```python
messages =  [  
{'role':'system', 'content':'You are friendly customer service telephone operator that speaks in Shakespearean English.'},    
{'role':'user', 'content':'Hi, my name is Liza'}  ]
response = get_output_from_messages(messages, temperature_input=1)
print(response)
```

```python
def initialize_chatbot(system_message):
    context = [ {'role':'system', 'content': system_message} ]
    response = get_output_from_messages(context, temperature_input = 1)
    context.append({'role':'assistant', 'content':f"{response}"})
    return context, response
```

```python
system_message = f"""
You are a tutor working with middle school students who are taking an online course and are very busy.
They struggle to manage their schedule and need help pacing themselves through their learning.
Your response should help the student in the form of a specific plan with timeline for learning a topic.
"""
```

```python
context, response = initialize_chatbot(system_message)
print(response)
```

```python
def add_new_message(context, user_message):
    context.append({'role':'user', 'content':f"{user_message}"})
    response = get_output_from_messages(context, temperature_input = 1)
    context.append({'role':'assistant', 'content':f"{response}"})
    return context, response
```

```python
system_message = f"""
You are a tutor for middle school students who are taking an online course and are very busy.
They struggle to manage their schedule and need help pacing themselves through their learning.
Ask simple questions, one at a time to help them organize their thoughts, and wait for their answers.
First, greet the student in a friendly way.
Then, gather information from them about their current studies.  
Make sure to ask them about the topic they are learning, their assignment timeline, and their available time for studying.
Finally, help the student create a specific plan with timeline for learning a topic.
"""

context, response = initialize_chatbot(system_message)
print(response)
```
