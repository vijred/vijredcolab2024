Everything in Python - every number, string, function, dataframe, etc. - is an **object**.   
Objects can be created several ways, the simplest is to assign them a value using the '=" operator.   
Scalar (single-value) objects can be numerical, strings, Boolean (True or False), dates or times.

Python **f-strings** (formated string literals) allow you to include expressions within your text, using curly braces{}.

```python
name = "Liza"
course = 478
sentence = f'Hi! my name is {name}.  In Spring B, I am teaching GBA {course}'
print(sentence)
```

A **list** is a multi-valued collection of Python objects.   
Lists are denoted by square brackets [] and the elements in a list are separated by commas.   
Lists have an **append** method, which allows items to be added to the list.   
List elements can be accessed by their index, singley or in a range using the 2:5 syntax (the colon is like a dash in the expression "elements 2 - 5".   
Note that the first index position of a list is 0.


```python
l = [1, True, "Opal", 3, 2, "Table"]
```

```python
print(l)
```

```python
type(l)
```

```python
l.append("new item")
```

```python
l[2]
```

```python
l[1:4]
```

A **dictionary** is a multi-valued collection of key-value pairs.   
Dictionaries are denoted by curly braces {} and the elements in a dictionary are separated by commas.  Colons separate keys and values for each pair.   
Dictionaries have a wide ranges of uses.   
Dictionary elements can be accessed and inserted using their keys.


```python
Sides = {"Triangle" : 3,"Square" : 4,"Pentagon" : 5,"Hexagon" : 6,"Heptagon" : 7,"Octagon" : 8,"Nonagon" : 9,"Decagon" : 10}
```

```python
Sides.keys()
```

```python
Sides.values()
```

```python
Sides['Square']
```

```python
Sides['Dodecagon'] = 12
Sides
```

Pandas includes a DataFrame structure.  Pandas dataframes are two-dimensional objects with both a row and column index.  You can think of a Pandas dataframe as similar to a spreadsheet or table.   
Dataframes can be created many different ways, often by reading a .csv.   
Dataframe elements can be accessed using their index as well as with the **.loc** and **.iloc** operators.

```python
data = {'city' : ['Buffalo', 'Rochester', 'Syracuse', 'Troy', 'Albany', 'New York City'],
        'population' : [275352, 207537, 142110, 50453, 102988, 7931147],
        'mayor' : ['Byron Brown', 'Malik Evans', 'Ben Walsh', 'Carmella Montello', 'Kathy Sheehan', 'Eric Adams']}

df = pd.DataFrame(data)
df
```
